```python
import numpy as np
import os, sys
import matplotlib.pyplot as plt


def read_files(path):  

    length = 1000
    N = 4
    x, y, z = np.zeros((N,length)), np.zeros((N,length)), np.zeros((N,length))
    vx, vy, vz = np.zeros((N,length)), np.zeros((N,length)), np.zeros((N,length))
    t = np.zeros(length)

    file_counter = 0
    for filename in os.listdir(path):
        with open(path + filename, 'r') as infile:
            lines = infile.readlines()[2:]
            line_counter = 0
            for line in lines:
                line = line.split()
                x[file_counter, line_counter] = float(line[1])
                y[file_counter, line_counter] = float(line[2])
                z[file_counter, line_counter] = float(line[3])
                vx[file_counter, line_counter] = float(line[4])
                vy[file_counter, line_counter] = float(line[5])
                vz[file_counter, line_counter] = float(line[6])
                line_counter += 1
            file_counter += 1
    
    # N

    return x,y,z,vx,vy,vz

mypath                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5D/"

x_,y_,z_, vx_, vy_, vz_ = read_files(mypath)
```
