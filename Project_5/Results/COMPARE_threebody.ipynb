```python
import numpy as np 
import os
import seaborn as sns

from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
%matplotlib inline

mypath0                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/ThreeBody/Planet_Sun_com.txt"
mypath1                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/ThreeBody/Planet_Earth_com.txt"
mypath2                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/ThreeBody/Planet_Jupiter_com.txt"

mypath3                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/ThreeBody/Planet_Sun_fix.txt"
mypath4                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/ThreeBody/Planet_Earth_fix.txt"
mypath5                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/ThreeBody/Planet_Jupiter_fix.txt"


#pathlist = np.array([mypath0,mypath1,mypath2,mypath3,mypath4,mypath5,mypath6,mypath7,mypath8])
pathlist = np.array([mypath0, mypath1, mypath2, mypath3, mypath4, mypath5])
planetnames = np.array(['Sun CoM', 'Earth CoM', 'Jupiter CoM', 'Sun Fix', 'Earth Fix', 'Jupiter Fix'])
fsize = 30
def read_files(path):  
    
    x, y, z = [],[],[]
    vx, vy, vz = [],[],[]
    
    

    with open(path, 'r') as infile:
        lines = infile.readlines()[2:]
        for line in lines:

            line = line.split()
            x.append(float(line[1]))
            y.append(float(line[2]))
            z.append(float(line[3]))
            vx.append(float(line[4]))
            vy.append(float(line[5]))
            vz.append(float(line[6]))
    
    r = np.zeros((len(x), 3))
    v = np.zeros((len(x), 3))
    for i in range(len(x)):
        r[i] = [x[i], y[i], z[i]]
        v[i] = [vx[i], vy[i], vz[i]]
        
    return r, v


fig = plt.figure(figsize = (20,15))
ax = plt.axes(projection='3d')

idx = -1


#ax.scatter3D(0,0,0, s = 50, c = 'orange', label = 'The sun')


i = 0
r0, v0 = read_files(pathlist[i])
xline = r0[:idx,0]
yline = r0[:idx,1]
zline = r0[:idx,2]
ax.plot3D(xline, yline, zline, ms = 25,label = planetnames[i], linewidth = 2)

i = 1
r0, v0 = read_files(pathlist[i])
xline = r0[:idx,0]
yline = r0[:idx,1]
zline = r0[:idx,2]
ax.plot3D(xline, yline, zline, ms = 25,label = planetnames[i], linewidth = 2)


i = 3
r0, v0 = read_files(pathlist[i])
xline = r0[:idx,0]
yline = r0[:idx,1]
zline = r0[:idx,2]
ax.plot3D(xline, yline, zline , '.', ms = 25,label = planetnames[i], linewidth = 2, c = 'y')

i = 4
r0, v0 = read_files(pathlist[i])
xline = r0[:idx,0]
yline = r0[:idx,1]
zline = r0[:idx,2]
ax.plot3D(xline, yline, zline, ms = 25,label = planetnames[i], linewidth = 2)



#ax.set_xlim([-2,2])
#ax.set_ylim([-2,2])
#ax.set_zlim([-1,1])

plt.title('Three body simulation -  Earth', pad = 50, fontsize = fsize+ 10)
ax.set_xlabel('X [AU]', fontsize = fsize)
ax.set_ylabel('Y [AU]',  fontsize = fsize)
#ax.set_zlabel('Z [AU]',  fontsize = fsize)
ax.xaxis.set_tick_params(labelsize=fsize-10, pad = 15)
ax.yaxis.set_tick_params(labelsize=fsize-10, pad = 15)
ax.zaxis.set_tick_params(labelsize=fsize-10, pad = 25)

plt.legend(fontsize = fsize-10)

ax.yaxis.labelpad=50
ax.xaxis.labelpad=40
#ax.zaxis.labelpad= 60


plt.savefig('../Paper/Figures/Comparing_Sun_Earth.png')
```
