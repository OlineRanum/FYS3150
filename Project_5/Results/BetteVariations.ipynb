```python
import numpy as np 
import os
import seaborn as sns
import matplotlib.pyplot as plt

mypathVel2                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5D4/EscapeVel2.txt"
mypathVel21                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5D4/EscapeVel2.1.txt"
mypathVel22                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5D4/EscapeVel2.2.txt"
mypathVel23                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5D4/EscapeVel2.3.txt"
mypathVel24                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5D4/EscapeVel2.4.txt"
mypathVel25                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5D4/EscapeVel2.5.txt"
mypathVel26                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5D4/EscapeVel2.6.txt"
mypathVel27                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5D4/EscapeVel2.7.txt"
mypathVel28                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5D4/EscapeVel2.8.txt"
mypathVel29                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5D4/EscapeVel2.9.txt"
mypathVel3                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5D4/EscapeVel3.txt"





def read_files(path):  
    
    x, y, z = [],[],[]
    vx, vy, vz = [],[],[]
    t = []
    
    
    with open(path, 'r') as infile:
        lines = infile.readlines()[2:]
        for line in lines:

            line = line.split()
            t.append(float(line[0]))
            x.append(float(line[1]))
            y.append(float(line[2]))
            z.append(float(line[3]))
            vx.append(float(line[4]))
            vy.append(float(line[5]))
            vz.append(float(line[6]))
    
    r = np.zeros((len(x), 3))
    v = np.zeros((len(x), 3))
    for i in range(len(x)):
        r[i] = [x[i], y[i], z[i]]
        v[i] = [vx[i], vy[i], vz[i]]
        
    return t, r, v

tE2, rE2, vE2 = read_files(mypathVel2)
tE21, rE21, vE21 = read_files(mypathVel21)
tE22, rE22, vE22 = read_files(mypathVel22)
tE23, rE23, vE23 = read_files(mypathVel23)
tE24, rE24, vE24 = read_files(mypathVel24)
tE25, rE25, vE25 = read_files(mypathVel25)
tE26, rE26, vE26 = read_files(mypathVel26)
tE27, rE27, vE27 = read_files(mypathVel27)
tE28, rE28, vE28 = read_files(mypathVel28)
tE29, rE29, vE29 = read_files(mypathVel29)
tE3, rE3, vE3 = read_files(mypathVel3)

c1 = 'firebrick'
c2 = 'steelblue'
c3 = 'plum'
c4 = 'seagreen'
c5 = 'orange'
```
