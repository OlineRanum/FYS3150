```python
import numpy as np 
import os
import seaborn as sns
import matplotlib.pyplot as plt

mypath1                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Earth.txt"
mypath2                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Jupiter.txt"

def read_files(path):  
    
    x, y, z = [],[],[]
    vx, vy, vz = [],[],[]
    
    

    with open(path, 'r') as infile:
        lines = infile.readlines()[2:]
        for line in lines:

            line = line.split()
            x.append(float(line[0]))
            y.append(float(line[1]))
            z.append(float(line[2]))
            vx.append(float(line[3]))
            vy.append(float(line[4]))
            vz.append(float(line[5]))
    
    r = np.zeros((len(x), 3))
    v = np.zeros((len(x), 3))
    for i in range(len(x)):
        r[i] = [x[i], y[i], z[i]]
        v[i] = [vx[i], vy[i], vz[i]]
        
    return r, v

r1, v1 = read_files(mypath1)
r2, v2 = read_files(mypath2)
```

```python
from mpl_toolkits import mplot3d


fig = plt.figure(figsize = (20,10))
ax = plt.axes(projection='3d')



# Data for a three-dimensional line
xline = r1[:,0]
zline = r1[:,1]
yline = r1[:,2]
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
xdata = r1[:,0]
zdata = r1[:,1]
ydata = r1[:,2]

ax.scatter3D(xdata, ydata, zdata, marker = 'o');


ax.scatter3D(0,0,0, s = 50, c = 'y')


# Data for a three-dimensional line
xline = r2[:,0]
zline = r2[:,1]
yline = r2[:,2]
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
xdata = r2[:,0]
zdata = r2[:,1]
ydata = r2[:,2]

ax.scatter3D(xdata, ydata, zdata, marker = 'o');


ax.scatter3D(0,0,0, s = 50, c = 'y')


ax.set_xlabel('X', fontsize = 20)
ax.set_ylabel('Y',  fontsize = 20)
ax.set_zlabel('Z',  fontsize = 20)
ax.xaxis.set_tick_params(labelsize=15, pad = 10)
ax.yaxis.set_tick_params(labelsize=15, pad = 10)
ax.zaxis.set_tick_params(labelsize=15, pad = 10)


ax.yaxis.labelpad=30
ax.xaxis.labelpad=30
ax.zaxis.labelpad=30
```
