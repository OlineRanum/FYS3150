```python
import numpy as np 
import os
import seaborn as sns

from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
%matplotlib

mypath0                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Sun.txt"
mypath1                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Mercury.txt"
mypath2                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Venus.txt"
mypath3                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Earth.txt"
mypath4                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Mars.txt"
mypath5                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Jupiter.txt"
mypath6                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Saturn.txt"
mypath7                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Uranus.txt"
mypath8                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Netptun.txt"

pathlist = np.array([mypath0,mypath1,mypath2,mypath3,mypath4,mypath5,mypath6,mypath7,mypath8])

def read_files(path):  
    
    x, y, z = [],[],[]
    vx, vy, vz = [],[],[]
    
    

    with open(path, 'r') as infile:
        lines = infile.readlines()[2:]
        for line in lines:

            line = line.split()
            x.append(float(line[0]))
            y.append(float(line[1]))
            z.append(float(line[2]))
            vx.append(float(line[3]))
            vy.append(float(line[4]))
            vz.append(float(line[5]))
    
    r = np.zeros((len(x), 3))
    v = np.zeros((len(x), 3))
    for i in range(len(x)):
        r[i] = [x[i], y[i], z[i]]
        v[i] = [vx[i], vy[i], vz[i]]
        
    return r, v


fig = plt.figure(figsize = (20,10))
ax = plt.axes(projection='3d')

idx = -1


for i in range(9):
    print(pathlist[i])
    r0, v0 = read_files(pathlist[i])
    xline = r0[:idx,0]
    yline = r0[:idx,1]
    zline = r0[:idx,2]
    ax.plot3D(xline, yline, zline)


ax.set_xlabel('X', fontsize = 20)
ax.set_ylabel('Y',  fontsize = 20)
ax.set_zlabel('Z',  fontsize = 20)
ax.xaxis.set_tick_params(labelsize=15, pad = 10)
ax.yaxis.set_tick_params(labelsize=15, pad = 10)
ax.zaxis.set_tick_params(labelsize=15, pad = 10)

ax.set_xlim([-2,2])
ax.set_ylim([-2,2])
ax.set_zlim([-2,2])


ax.yaxis.labelpad=30
ax.xaxis.labelpad=30
ax.zaxis.labelpad=30
```
