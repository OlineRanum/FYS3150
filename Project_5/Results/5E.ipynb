```python
import numpy as np 
import os
import seaborn as sns

from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
%matplotlib inline

mypath0                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Sun.txt"
#mypath1                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Mercury.txt"
#mypath2                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Venus.txt"
mypath3                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Earth_NormalM.txt"
mypath3_NoJ                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Earth_NoJ.txt"
mypath3_m10                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Earth10M.txt"
mypath3_m1000                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Earth_1000M.txt"

#mypath4                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Mars.txt"
mypath5                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Jupiter_NormalM.txt"
#mypath6                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Saturn.txt"
#mypath7                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Uranus.txt"
#mypath8                  = "/home/oline/Documents/UiO/FYS3150/FYS3150/Project_5/Results/5E/Planet_Netptun.txt"

#pathlist = np.array([mypath0,mypath1,mypath2,mypath3,mypath4,mypath5,mypath6,mypath7,mypath8])
pathlist = np.array([mypath3, mypath3_NoJ, mypath3_m10, mypath3_m1000])
planetnames = np.array(['Earth with Jupiter','Erath without Jupiter', 'Earth with M$_{Jupiter}$x10', 'Earth with M$_{Jupiter}$x1000'])
fsize = 25
def read_files(path):  
    
    x, y, z = [],[],[]
    vx, vy, vz = [],[],[]
    
    

    with open(path, 'r') as infile:
        lines = infile.readlines()[2:]
        for line in lines:

            line = line.split()
            x.append(float(line[1]))
            y.append(float(line[2]))
            z.append(float(line[3]))
            vx.append(float(line[4]))
            vy.append(float(line[5]))
            vz.append(float(line[6]))
    
    r = np.zeros((len(x), 3))
    v = np.zeros((len(x), 3))
    for i in range(len(x)):
        r[i] = [x[i], y[i], z[i]]
        v[i] = [vx[i], vy[i], vz[i]]
        
    return r, v


fig = plt.figure(figsize = (20,15))
ax = plt.axes(projection='3d')

idx = -1


ax.scatter3D(0,0,0, s = 50, c = 'orange', label = 'The sun')


i = 0
r0, v0 = read_files(pathlist[i])
xline = r0[:idx,0]
yline = r0[:idx,1]
zline = r0[:idx,2]
ax.plot3D(xline, yline, zline, ms = 25,label = planetnames[i], linewidth = 2)

i = 1
r0, v0 = read_files(pathlist[i])
xline = r0[:idx,0]
yline = r0[:idx,1]
zline = r0[:idx,2]
ax.plot3D(xline, yline, zline, ms = 25,label = planetnames[i], linewidth = 2)
"""
i = 2
r0, v0 = read_files(pathlist[i])
xline = r0[:idx,0]
yline = r0[:idx,1]
zline = r0[:idx,2]
ax.plot3D(xline, yline, zline, ms = 25,label = planetnames[i], linewidth = 2)

i = 3
idx = 800
r0, v0 = read_files(pathlist[i])
xline = r0[:-idx,0]
yline = r0[:-idx,1]
zline = r0[:-idx,2]
ax.plot3D(xline, yline, zline, ms = 25,label = planetnames[i], linewidth = 2)
"""


#ax.set_xlim([-2,2])
#ax.set_ylim([-2,2])
#ax.set_zlim([-2,2])

plt.title('Sun-Earth-Jupiter system', pad = 50, fontsize = fsize+ 10)
ax.set_xlabel('X [AU]', fontsize = fsize)
ax.set_ylabel('Y [AU]',  fontsize = fsize)
ax.set_zlabel('Z [AU]',  fontsize = fsize)
ax.xaxis.set_tick_params(labelsize=fsize-10, pad = 15)
ax.yaxis.set_tick_params(labelsize=fsize-10, pad = 15)
ax.zaxis.set_tick_params(labelsize=fsize-10, pad = 15)

plt.legend(fontsize = fsize-10)

ax.yaxis.labelpad=50
ax.xaxis.labelpad=40
ax.zaxis.labelpad=40


plt.savefig('../Paper/Figures/SEJ_Part1.png')
```
