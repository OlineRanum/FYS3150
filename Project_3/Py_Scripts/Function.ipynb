```python
import numpy as np 
import matplotlib.pyplot as plt

limits = 400
x = np.linspace(-limits,limits,int(1e4))
y = z = x
r = np.sqrt(x**2+y**2+z**2)
```

```python
def distribution(x,y,z):
    alpha = 2
    x1, y1, z1 = 0, 0, 0
    return np.exp(-2*alpha*(np.sqrt(x1**2+y1**2+z1**2) + x**2+y**2+z**2))/np.sqrt((x1-x)**2+(y1-y)**2 + (z1-z**2))

func = np.zeros(len(x))
for i in range(len(x)):
    func[i] = distribution(x[i],y[i],z[i])
```

```python
plt.figure(figsize = (20,10))
plt.plot(x,func)
plt.ylim([0,1])
plt.xlim([-2,2])
```
