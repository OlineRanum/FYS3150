```python
import numpy as np
import matplotlib.pyplot as plt

exact = 0.19276571

N = np.array([5,11,15,19,21,23,25,27,29,31,33,35,37,39])
vals = np.array([0.264249, 0.291261, 0.239088, 0.211832, 0.204307, 0.199232, 0.195817, 0.193524, 0.191995, 0.190985, 0.190331, 0.189923, 0.189684, 0.189562])
```

```python
fsize = 40
plt.figure(figsize = (15,10))


plt.title('Integration values by Gauss-Legendre quadrature\n',fontsize = fsize)
plt.yticks(fontsize = fsize-5)
plt.xticks(fontsize = fsize-5)


plt.plot(N, vals, '.', c = 'r', ms = 20)
plt.plot([N[0],N[-1]], [exact, exact])



plt.legend(['I$_{G-LEG}$', 'Exact value'], fontsize = fsize)
#plt.ylim([-5,250])
plt.xlabel('N [#]',  fontsize = fsize, labelpad = 25)
plt.ylabel('I$_{Gaus-Legendre}$', fontsize = fsize,  labelpad = 50)
#plt.xlim([-2.2,2.2])
plt.savefig('Gauss_Legendre.png', bbox_inches = 'tight')
```
