```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.colors

limits = 5
x = np.linspace(-limits,limits,int(1e4))
y = z = x
r = np.sqrt(x**2+y**2+z**2)
```

```python
def distribution(x,y,z):
    alpha = 2
    x1, y1, z1 = 0, 0, 0
    return np.exp(-2*alpha*(np.sqrt(x1**2+y1**2+z1**2) + x**2+y**2+z**2))#/np.sqrt((x1-x)**2+(y1-y)**2 + (z1-z**2))

func = np.zeros(len(x))
for i in range(len(x)):
    func[i] = distribution(x[i],y[i],z[i])
```

```python
fsize = 40
plt.figure(figsize = (15,10))
plt.title('$\Psi$(r$_1$,r$_2)$\n',fontsize = fsize)
plt.yticks(fontsize = fsize-5)
plt.xticks([-2.0, -1.0, 0 ,1.0,2.0], fontsize = fsize-5)

plt.plot(x,func, c = 'cornflowerblue', linewidth = 3)


plt.legend(['$\Psi$(r$_1$,r$_2)$'], fontsize = fsize)
#plt.ylim([-5,250])
plt.xlabel('r = r$_1$-r$_2$',  fontsize = fsize, labelpad = 25)
plt.ylabel('$\Psi$(r$_1$,r$_2)$', fontsize = fsize,  labelpad = 50)
plt.xlim([-2.2,2.2])
plt.savefig('Wavefunction.png', bbox_inches = 'tight')
```
